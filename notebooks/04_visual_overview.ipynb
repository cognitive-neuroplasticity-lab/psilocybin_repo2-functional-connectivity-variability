```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
summary_df = pd.DataFrame({
    "Subject": ["sub-P1", "sub-P2", "sub-P3"],
    "Mean Connectivity": [0.138, 0.272, 0.096]
})

group_mean = summary_df["Mean Connectivity"].mean()
summary_df
```

```python
summary_df["Deviation"] = summary_df["Mean Connectivity"] - group_mean

plt.figure(figsize=(7,4))
sns.barplot(data=summary_df, x="Subject", y="Deviation")
plt.axhline(0, linestyle="--")
plt.title("Deviation from Group Mean Connectivity")
plt.tight_layout()
plt.show()
```

```python
min_val = summary_df["Mean Connectivity"].min()
max_val = summary_df["Mean Connectivity"].max()

plt.figure(figsize=(8,2))
plt.hlines(y=1, xmin=min_val, xmax=max_val, linewidth=3)
plt.scatter(summary_df["Mean Connectivity"], [1]*len(summary_df), s=100)

for _, row in summary_df.iterrows():
    plt.text(row["Mean Connectivity"], 1.02, row["Subject"], ha="center")

plt.yticks([])
plt.xlabel("Mean Connectivity")
plt.title("Observed Connectivity Range Across Subjects")
plt.tight_layout()
plt.show()
```

```python
ranked = summary_df.sort_values("Mean Connectivity", ascending=False).reset_index(drop=True)
ranked["Gap From Previous"] = ranked["Mean Connectivity"].diff().fillna(0)

ranked
```

```python
roadmap = pd.DataFrame({
    "Stage": [
        "3 Subjects",
        "All Subjects",
        "Multiple Sessions",
        "Condition Comparisons",
        "Predictive Models"
    ],
    "Progress": [1,2,3,4,5]
})

plt.figure(figsize=(8,4))
plt.plot(roadmap["Stage"], roadmap["Progress"], marker="o")
plt.xticks(rotation=30, ha="right")
plt.title("Planned Scaling Path for Repo 2")
plt.ylabel("Next Step Order")
plt.tight_layout()
plt.show()
```
